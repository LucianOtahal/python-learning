```python
import json
import requests
response = requests.get('http://api.nbp.pl/api/exchangerates/tables/a/') # ➡️ Posílá HTTP GET požadavek na veřejné API Národní banky Polska (NBP), 
# které vrací tabulku kurzů měn.

data = response.json() #➡️ Z odpovědi si vezme obsah ve formátu JSON a převede ho na Python objekt (většinou seznam nebo slovník).


print(json.dumps(data, indent=4, sort_keys=True)) # ➡️ Vypíše ten objekt jako formátovaný JSON text:
# - indent=4 → odsazení 4 mezery (přehlednější výstup)
# - sort_keys=True → klíče ve slovníku se seřadí abecedně
```
